```python
import numpy as np
import matplotlib.pyplot as plt

def logLoss(val,target):
    if target == 1:
        return -np.log(val)
    elif target == 0:
        return -np.log(1-val)
```

```python
x = np.arange(0,1, .02)
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(121)
plt.plot(x,logLoss(x, 1))
plt.title("Log loss, target = 1", fontsize=20)
plt.xlabel("$\hat{y}$", fontsize=15)
plt.ylabel("Loss", fontsize=15)
ax = fig.add_subplot(122)
plt.plot(x,logLoss(x, 0))
plt.title("Log loss, target = 0", fontsize=20)
plt.xlabel("$\hat{y}$", fontsize=15)
plt.ylabel("Loss", fontsize=15)
plt.savefig("logLossPlot.png", bbox_inches="tight")
plt.show()
```

```python
x=np.arange(0, 1, .02)
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(121)
plt.plot(x, logLoss(x,1))
plt.title("Log loss, target = 1", fontsize=20)
plt.xlabel("$\hat{y}$", fontsize=15)
plt.ylabel("Loss", fontsize=15)
plt.scatter(0.2, 1.6, color="orange", s=75)
plt.arrow(0.2, 1.5, dx=0.1, dy=-0.4, head_width=.05, width = 0.02, color="orange")
plt.text(0.1, 3, s="worse", color="red", fontsize=15)
plt.text(0.9, 0.3, s="better", color="green", fontsize=15)

ax = fig.add_subplot(122)
plt.plot(x, logLoss(x,0))
plt.title("Log loss, target = 0", fontsize=20)
plt.xlabel("$\hat{y}$", fontsize=15)
plt.ylabel("Loss", fontsize=15)
plt.scatter(0.8, 1.6, color="orange", s=75)
plt.arrow(0.8, 1.5, dx=-0.1, dy=-0.4, head_width=.05, width = 0.02, color="orange")
plt.text(0.8, 3, s="worse", color="red",fontsize=15)
plt.text(0.0, 0.3, s="better", color="green", fontsize=15)
plt.savefig("logLossMinimzePlot.png", bbox_inches="tight")
plt.show()
```
